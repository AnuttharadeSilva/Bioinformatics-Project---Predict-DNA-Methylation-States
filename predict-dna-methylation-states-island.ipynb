```python
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
```

```python
train='/kaggle/input/predict-dna-methylation/train.csv'
df=pd.read_csv(train,index_col=0)
df.head()
```

```python
X = pd.get_dummies(df['Relation_to_UCSC_CpG_Island'])
X['Beta']=df.Beta
X.head()
```

```python
X['Beta']=df.Beta
X.groupby('Island').mean()['Beta']
```

```python
X=X.drop('Beta',1)
```

```python
y=df.Beta
clf = LogisticRegression(random_state=0).fit(X, y)
y=y.values
prob=clf.predict_proba(X)
print('AUC: '+str(roc_auc_score(y,prob[:,1])))
pred=clf.predict(X)
print(classification_report(y, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
y=df.Beta
clf = RandomForestClassifier(max_depth=2, random_state=0).fit(X, y)
y=y.values
prob=clf.predict_proba(X)
print('AUC: '+str(roc_auc_score(y,prob[:,1])))
pred=clf.predict(X)
print(classification_report(y, pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
y=df.Beta
clf = KNeighborsClassifier(n_neighbors=3).fit(X, y)
y=y.values
prob=clf.predict_proba(X)
print('AUC: '+str(roc_auc_score(y,prob[:,1])))
pred=clf.predict(X)
print(classification_report(y, pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
y=df.Beta
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0).fit(X, y)
y=y.values
prob=clf.predict_proba(X)
print('AUC: '+str(roc_auc_score(y,prob[:,1])))
pred=clf.predict(X)
print(classification_report(y, pred))
```
